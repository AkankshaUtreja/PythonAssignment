Top 10 populated states of India

```python
color = ['magenta','red','green','blue','cyan','black','orange']
grouping_by_state_population = indian_cities[['population_total','state_name']].groupby('state_name').sum().sort(['population_total'],ascending=False)
grouping_by_state_population.plot(kind="bar", color = color) 
plt.xlabel('State Name')
plt.ylabel('Toal Population')
plt.show()
```

# Top 10 populated cities of India

```python

most_pop = indian_cities.sort('population_total',ascending=False)
top10_populated=most_pop.head(10)
top10_populated

## From the below we can see that Greater Mumbai is the most populated city in India.
```

```python
# Finding the latitue and longitute from the location.

indian_cities['latitude'] = indian_cities['location'].apply(lambda y: y.split(',')[0])
indian_cities['longitude'] = indian_cities['location'].apply(lambda y: y.split(',')[1])
indian_cities.head(1)
```

```python
## We can also plot this on India's map using basemap.

plt.subplots(figsize=(20, 15))
map = Basemap(width=1200000,height=900000,projection='lcc',resolution='l',
                    llcrnrlon=67,llcrnrlat=5,urcrnrlon=99,urcrnrlat=37,lat_0=28,lon_0=77)

map.drawmapboundary(fill_color='lightblue')
map.fillcontinents(color='white')
map.drawcountries(color='black')
map.drawcoastlines(linewidth=0.5,color='black')  

longi=array(top10_populated['longitude'])
lati=array(top10_populated['latitude'])
poptotal=array(top10_populated['population_total'])
namecity=array(top10_populated['name_of_city'])

x, y = map(longi, lati)
plt.plot(x, y, 'ro', markersize=10)


for ncs,xpoptotal, ypoptotal in zip(namecity, x, y):
    plt.text(xpoptotal+60000, ypoptotal+30000, ncs, fontsize=10, fontweight='bold')

plt.title('Top 10 Populated Cities in India',fontsize=20)
```

Literacy Analysis of India

```python
StateNameGroup = indian_cities.groupby('state_name')
y = StateNameGroup.median().reset_index()
plt.rcParams['figure.figsize'] = (6.0, 5.0)
genre_count = sns.barplot(y='state_name', x='effective_literacy_rate_total', data=y, palette="Greens", ci=None)
```

```python
# From the above graph, it is difficult to predict which state has highest literacy. 
#Let's try to find the top 10 literate cities in India.
```

Top 10 Literate cities of India

```python
print("Top 10 literate cities in India")
lit_top10 = indian_cities.sort('effective_literacy_rate_total',ascending=False).head(10)
lit_top10
```

```python
## We can use the basemap function to show these.
plt.subplots(figsize=(20, 15))
map = Basemap(width=1200000,height=900000,projection='lcc',resolution='l',
                    llcrnrlon=67,llcrnrlat=5,urcrnrlon=99,urcrnrlat=37,lat_0=28,lon_0=77)

map.drawmapboundary(fill_color='lightblue')
map.fillcontinents(color='white')
map.drawcountries(color='black')
map.drawcoastlines(linewidth=0.5,color='black')  

longi=array(lit_top10['longitude'])
lati=array(lit_top10['latitude'])
poptotal=array(lit_top10['population_total'])
nc=array(lit_top10['name_of_city'])

x, y = map(longi, lati)
plt.plot(x, y, 'ro', markersize=10)


for ncs, xpoptotal, ypoptotal in zip(nc, x, y):
    plt.text(xpoptotal+60000, ypoptotal+30000, ncs, fontsize=10, fontweight='bold')

plt.title('Top 10 Literate Cities in India',fontsize=20)

```

Analysing the Sex Ratio in India

```python
color = ['pink','red','green','blue','cyan','black','orange']
grouping_by_sexratio = indian_cities[['sex_ratio','state_name']].groupby('state_name').mean().sort(['sex_ratio'],ascending=True)
grouping_by_sexratio.plot(kind="bar", color = color) 
plt.xlabel('State Name')
plt.ylabel('Sex Ratio')
plt.show()
```

* The literacy rate analysis is done which shows that the most literate city is Aizawl.
