Let's try to find the places where earthquakes took place

```python
# Let's try to find the type of eathquakes that are present in the dataset.
earthquake_recent=pd.DataFrame()
earthquake_recent=earthquake1[earthquake1['Type']=='Earthquake']
earthquake_recent
```

```python
earthquake_recent=pd.DataFrame()
earthquake_recent=earthquake1[earthquake1['Type']!='Earthquake']
earthquake_recent
```

```python
# Let's try to show this on map.

mapp = Basemap(projection='mill',llcrnrlat=-80,urcrnrlat=80, llcrnrlon=-180,urcrnrlon=180,lat_ts=20,resolution='c')
fig = plt.figure(figsize=(12,10))



longitudes_rec2 = earthquake_recent["Longitude"].tolist()
latitudes_rec2 = earthquake_recent["Latitude"].tolist()

a,b= mapp(longitudes_rec2,latitudes_rec2)

plt.title("The red dots are the places where earthquakes occured")
mapp.plot(a, b, "o", markersize = 3, color = 'red')

mapp.drawcoastlines()
mapp.drawcountries()
mapp.fillcontinents(color='pink',lake_color='blue')
mapp.drawmapboundary()
mapp.drawcountries()
plt.show()
```

Let's find out the highly affected areas

```python
## Let's suppose the earthquakes which have magnitute greater than 8 are high intensity earthquakes and let's try to find 
## the highly affected areas.

highlyaffected = earthquake1[earthquake1["Magnitude"]>=8]
highlyaffected
```

```python
## Plotting the map for the highly affected areas to get clarity.

longi = highlyaffected["Longitude"].tolist()
lati = highlyaffected["Latitude"].tolist()
mapp = Basemap(projection='mill',llcrnrlat=-80,urcrnrlat=80, llcrnrlon=-180,urcrnrlon=180,lat_ts=20,resolution='c')
x,y = mapp(longi,lati)
fig3 = plt.figure(3,figsize=(12,10))
plt.title("Highly earthquake affected areas")
mapp.plot(x, y, "o", markersize = 10, color = 'red')
mapp.drawcoastlines()
mapp.fillcontinents(color='green',lake_color='aqua')
mapp.drawmapboundary()
mapp.drawcountries()
plt.show()
```

Overall Magnitute of Earthquakes

```python
## Let's try to find the magnitute of the earthquakes by finding the minimum, maximum and average.

mini = earthquake1["Magnitude"].min()
maxi = earthquake1["Magnitude"].max()
avg = earthquake1["Magnitude"].mean()

print("Minimum:", mini)
print("Maximum:",maxi)
print("Mean",avg)
```

```python
## Plotting the graph for the above.

(no,bins, patches) = plt.hist(earthquake1["Magnitude"], range=(0,10), bins=10)
plt.xlabel("Magnitute of Earthquakes")
plt.ylabel("Number of Occurences")
plt.title("Representation of Earthquake magnitutes")

print("Magnitude" +"   "+ "Number of Occurence")
for i in range(5, len(no)):
    print(str(i)+ "-"+str(i+1)+"         " +str(no[i]))
```

Relation between Magnitute and Depth

```python
## Plotting the scatter plot for finding the relation between Magnitute and depth.

plt.scatter(earthquake1["Magnitude"],earthquake1["Depth"])
```
